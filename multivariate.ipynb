```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = pd.read_csv("https://raw.githubusercontent.com/koroteevmv/ML_course/refs/heads/main/ML1.2%20multivariate/data/2_x.csv",
                header=None)
y = pd.read_csv("https://raw.githubusercontent.com/koroteevmv/ML_course/refs/heads/main/ML1.2%20multivariate/data/2_y.csv",
                header=None)
```

```python
print("Размер X:", x.shape)
print("Размер Y:", y.shape)
```

```python
n_features = x.shape[1]
print(f"Всего признаков: {n_features}")
```

```python
for i in range(n_features):
    print(f"\n--- Анализ признака {i} ---")

    X_i = x[[i]]

    model = LinearRegression()
    model.fit(X_i, y)

    coef = model.coef_[0][0]
    intercept = model.intercept_[0]
    print(f"Коэффициент наклона: {coef:.4f}")
    print(f"Свободный член: {intercept:.4f}")

    r2 = model.score(X_i, y)
    print(f"R² = {r2:.4f}")

    plt.figure(figsize=(8, 5))

    # Точки исходных данных
    plt.scatter(x[i], y, alpha=0.5, label='Данные')

    # Линия регрессии
    xx = np.linspace(x[i].min(), x[i].max(), 100)
    xx_2d = xx.reshape(-1, 1)
    yy = model.predict(xx_2d)
    plt.plot(xx, yy, c='red', linewidth=2, label='Линия регрессии')

    # Оформление
    plt.xlabel(f'Признак {i}')
    plt.ylabel('Целевая переменная')
    plt.title(f'Парная регрессия: признак {i} (R² = {r2:.4f})')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
```

ВЫВОД ПО РЕЗУЛЬТАТАМ:
- Лучший признак: Признак 2 (R² = 0.4000)
Объясняет 40% вариации целевой переменной. Умеренная положительная связь.

- Худший признак: Признак 3 (R² = 0.1044)
Объясняет всего 10% вариации. Очень слабая связь.

Общее заключение:
- Ни один из признаков по отдельности не дает хорошего предсказания (максимальный R² = 0.4). Это говорит о том, что целевая переменная зависит от комбинации признаков, поэтому множественная регрессия (со всеми признаками) должна показать значительно лучший результат.

2. (*) Модифицируйте код из предыдущей работы так, чтобы модель регрессии умела работать с входными данными, состоящими из произвольного количества столбцов.


```python
class MultipleRegression:
    """Модель множественной линейной регрессии"""

    def __init__(self):
        # Коэффициенты будут инициализированы при обучении
        self.b = None

    def predict(self, X):
        """Предсказание: y = b0 + b1*x1 + b2*x2 + ..."""
        X = np.array(X)
        # Добавляем столбец единиц для b0
        X_with_ones = np.hstack([np.ones((X.shape[0], 1)), X])
        return X_with_ones @ self.b

    def error(self, X, Y):
        """Среднеквадратичная ошибка"""
        Y_pred = self.predict(X)
        Y = np.array(Y).flatten()
        return np.sum((Y_pred - Y) ** 2) / (2 * len(Y))

    def fit(self, X, Y, alpha=0.01, max_steps=1000):
        """Градиентный спуск для множественной регрессии"""
        X = np.array(X)
        Y = np.array(Y).flatten()
        m = len(Y)  # количество объектов
        n = X.shape[1]  # количество признаков

        # Добавляем столбец единиц для b0
        X_with_ones = np.hstack([np.ones((m, 1)), X])

        # Инициализируем коэффициенты нулями
        self.b = np.zeros(n + 1)

        # Для отслеживания ошибки
        errors = []

        for step in range(max_steps):
            # Предсказание
            Y_pred = X_with_ones @ self.b

            # Градиенты для всех коэффициентов сразу
            gradients = (1/m) * (X_with_ones.T @ (Y_pred - Y))

            # Обновляем коэффициенты
            self.b -= alpha * gradients

            # Сохраняем ошибку
            errors.append(self.error(X, Y))

        return errors

    def score(self, X, Y):
        """R² коэффициент детерминации"""
        Y_pred = self.predict(X)
        Y = np.array(Y).flatten()

        # Остаточная сумма квадратов
        rss = np.sum((Y - Y_pred) ** 2)
        # Общая сумма квадратов
        tss = np.sum((Y - np.mean(Y)) ** 2)

        return 1 - (rss / tss)
```

```python
# Шаг 4. Проверяем работу модели
print("\n" + "="*50)
print("ОБУЧЕНИЕ МОДЕЛИ МНОЖЕСТВЕННОЙ РЕГРЕССИИ")
print("="*50)

# Создаем модель
model = MultipleRegression()

# Обучаем на всех признаках
errors = model.fit(x, y, alpha=0.01, max_steps=1000)

# Выводим результаты
print(f"\nКоэффициенты модели:")
print(f"b0 (свободный член) = {model.b[0]:.4f}")
for i in range(1, len(model.b)):
    print(f"b{i} (признак {i-1}) = {model.b[i]:.4f}")

print(f"\nФинальная ошибка: {model.error(x, y):.4f}")
print(f"R² score: {model.score(x, y):.4f}")

```

```python
# Шаг 5. Визуализация обучения
plt.figure(figsize=(10, 4))

# График ошибки
plt.subplot(1, 2, 1)
plt.plot(errors)
plt.xlabel('Шаг')
plt.ylabel('Ошибка')
plt.title('Кривая обучения')
plt.grid(True)

# График предсказаний vs реальные значения
plt.subplot(1, 2, 2)
y_pred = model.predict(x)
plt.scatter(y, y_pred, alpha=0.5)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.xlabel('Реальные значения')
plt.ylabel('Предсказанные значения')
plt.title(f'Предсказания vs Реальность (R² = {model.score(x, y):.4f})')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Шаг 6. Сравнение с парной регрессией (для понимания)
print("\n" + "="*50)
print("СРАВНЕНИЕ С ПАРНОЙ РЕГРЕССИЕЙ")
print("="*50)

# Обучаем на одном признаке для сравнения
model_single = MultipleRegression()
errors_single = model_single.fit(x[[2]], y, alpha=0.01, max_steps=1000)

print(f"Парная регрессия (только признак 2):")
print(f"  R² = {model_single.score(x[[2]], y):.4f}")
print(f"Множественная регрессия (все признаки):")
print(f"  R² = {model.score(x, y):.4f}")
print(f"Улучшение: +{model.score(x, y) - model_single.score(x[[2]], y):.4f}")
```

Результаты:
